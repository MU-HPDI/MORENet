```python
District_ACT.head()
```

```python
District_ACT = District_ACT[District_ACT['ACT_TESTS_ADMINISTERED']!='*']
District_ACT
```

# Math and Statistics operations

```python
Boone_df = District_ACT[District_ACT['DISTRICT_NAME']=='BOONE CO. R-IV']
Boone_df
```

```python
Boone_df_ACT_TESTS_ADMINISTERED = Boone_df['ACT_TESTS_ADMINISTERED']
Boone_df_ACT_TESTS_ADMINISTERED
```

## min and max

```python
min_val = Boone_df_ACT_TESTS_ADMINISTERED.min()
max_val = Boone_df_ACT_TESTS_ADMINISTERED.max()

print('the min value: {}'.format(min_val))
print('the max value: {}'.format(max_val))
```

## mean

```python
mean_val = Boone_df_ACT_TESTS_ADMINISTERED.mean()
print('the mean value: {:3.2f}'.format(mean_val))
```

## Median

```python
median_val = Boone_df_ACT_TESTS_ADMINISTERED.median()
print('the median value: {:3.2f}'.format(median_val))
```

## quantiles

```python
Q25 = Boone_df_ACT_TESTS_ADMINISTERED.quantile(0.25)
Q50 = Boone_df_ACT_TESTS_ADMINISTERED.quantile(0.50)
Q75 = Boone_df_ACT_TESTS_ADMINISTERED.quantile(0.75) 

print('the first quantile value : {:3.2f}'.format(Q25))
print('the second quantile value: {:3.2f}'.format(Q50))
print('the third quantile value : {:3.2f}'.format(Q75))
```

# Plots

## Graph

```python
plt.plot(Boone_df['YEAR'],Boone_df_ACT_TESTS_ADMINISTERED)
plt.xlabel('years')
plt.ylabel('number of tests administered')
plt.title('Number of ACT tests administered in Bonne county')
```

```python
plt.plot(Boone_df['YEAR'],Boone_df_ACT_TESTS_ADMINISTERED,label='test administered')
plt.plot(Boone_df['YEAR'],Boone_df['GRADUATES'],label='graduates')
plt.xlabel('years')
plt.ylabel('number of tests administered')
plt.title('ACT tests administered VS Graduates in Bonne county')
plt.legend()
```
